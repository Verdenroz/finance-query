# finance-query-py quickstart

Native Python bindings for the finance-query Rust library. Async-first, type-safe, polars DataFrames.

## Install

```
pip install finance-query-py
```


## Quote a single ticker

````python
from finance_query import Ticker

ticker = await Ticker.new("AAPL")
quote = await ticker.quote()
print(f"{quote.symbol}: {quote.short_name}")
````

## Chart → polars DataFrame

````python
from finance_query import Interval, TimeRange

chart = await ticker.chart(Interval.OneDay, TimeRange.OneMonth)
df = chart.to_dataframe()
df.head()
````

## Batch multiple tickers

````python
from finance_query import Tickers

tickers = await Tickers.new(["AAPL", "MSFT", "NVDA"])
result = await tickers.quotes()
for sym, q in result.data.items():
    print(f"{sym}: {q.short_name}")
if result.errors:
    print("errors:", result.errors)
````

## Market search

````python
from finance_query import finance

results = await finance.search("Apple")
for r in results[:5]:
    print(r.symbol)
````

## Fear and Greed Index

````python
fng = await finance.fear_and_greed()
print(f"{fng.value}/100 — {fng.classification}")
````
